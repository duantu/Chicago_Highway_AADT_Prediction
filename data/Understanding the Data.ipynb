2. Question: Proj 2 and Proj 5 do not make any changes on #links or Capacity, but in sheet P2 and P5, AADT_b \neq AADT_a.
3. Proj 1 and Proj 6 share 15 links that are marked "Project" in both.

Current idea for the ML algorithm:
Given a graph, where each edge represent a road, the alg is given inputs of (1)#lanes and (2) maximum capacity of each edge and it outputs the AADT of each edge. 

```python
datadir = "../data/"
datafilename = "Certainty.xlsx"
datafilepath = datadir + datafilename
```

```python
from itertools import combinations
import pandas as pd
import numpy as np

df_p0= pd.read_excel(datafilepath, sheet_name='P0')
df_p0_unique = df_p0.drop_duplicates(subset=['Link ID']).dropna(subset=['Link ID', 'A X_COORD', 'A Y_COORD', 'B X_COORD', 'B Y_COORD'])
```

```python


# List of the first 6 integers
integers = [1, 2, 3, 4, 5, 6]

# Function to generate combinations
def generate_combinations(integers):
    all_combinations = []
    # Loop through lengths from 1 to 6
    for length in range(1, 7):
        # Generate combinations of the current length
        comb = combinations(integers, length)
        # Convert each combination to a string, add "p", and add to the list
        all_combinations.extend(['P' + ''.join(map(str, c)) for c in comb])
    return all_combinations

# Generate all combinations
sheet_names_arr = generate_combinations(integers)

# Insert "P0" at the beginning of the list
sheet_names_arr.insert(0, "P0")
```

```python
# Initialize a data frame to represent changes in #lanes before and after each project

df_change_num_lanes = pd.DataFrame(columns= sheet_names_arr, index = df_p0_unique['Link ID'])

# Calculate changes in #lanes for each project 
for ii in range(len(sheet_names_arr)):
    df= pd.read_excel(datafilepath, sheet_name=sheet_names_arr[ii]) 
    df_unique = df.drop_duplicates(subset=['Link ID']).dropna(subset=['Link ID', 'A X_COORD', 'A Y_COORD', 'B X_COORD', 'B Y_COORD'])

    df_change_num_lanes[sheet_names_arr[ii]] = df_unique["# of lanes-A"].sub(df_unique["# of lanes-B"]).values
    ii = ii+1
```

```python
df_change_num_lanes.columns[(df_change_num_lanes != 0).any()]

# Conclusion: P2 and P5 don't change #lanes for any link
```

```python
df_change_capacity = pd.DataFrame(columns= sheet_names_arr, index = df_p0_unique['Link ID'])

# Calculate changes in Capacity for each project 
for ii in range(len(sheet_names_arr)):
    df= pd.read_excel(datafilepath, sheet_name=sheet_names_arr[ii]) 
    df_unique = df.drop_duplicates(subset=['Link ID']).dropna(subset=['Link ID', 'A X_COORD', 'A Y_COORD', 'B X_COORD', 'B Y_COORD'])

    df_change_capacity[sheet_names_arr[ii]] = df_unique["Capacity-A (veh/h)"].sub(df_unique["Capacity-B (veh/h)"]).values
    ii = ii+1
```

```python
df_change_capacity.columns[(df_change_capacity != 0).any()]

# Conclusion: P2 and P5 don't change the capacity of any links 
```
